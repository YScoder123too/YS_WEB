# Joining dataframe with different columns

```python
dict3={
    'iid':[7,8,9],
    'cod':[38,40,42],
    'invoice_date':['2024-07-01','2024-07-02','2024-07-01'],
    'total':[1.98,2.56,3.6],
    'bill_city':['Panipat','Delhi','Gurugram']
}
```

```python
inv_feb=pd.DataFrame(dict3)
inv_feb
```

```python
pd.concat([inv_jan,inv_feb],sort=True)
```

```python
pd.concat([inv_jan,inv_feb],join='inner')
```

```python
tracks={
    'tid':[2,3,4],
    'name':['Balls to the...','Fast As a Shark','Restless and...'],
    'aid':[2,3,3],
    'mtid':[2,2,2],
    'gid':[1,1,1],
    'uprice':[0.99,0.99,0.99]    
}
```

```python
specs={
    'tid':[2,3,4],
    'milliseconds':[342562,3990994,43317739],
    'bytes':[232354,9735692,64512879]
}
```

```python
d1=pd.DataFrame(tracks)
d1
```

```python
d2=pd.DataFrame(specs)
d2
```

```python
pd.concat([d1,d2])
```

```python
pd.concat([d1,d2],sort=True)
```

```python
pd.concat([d1,d2],join='inner')
```

```python
pd.concat([d1,d2],join='outer')
```

```python
pd.concat([d1,d2],keys=['tracks','specs'],ignore_index=False,sort=True)
```

```python
d1.merge(d2,on='tid',validate='one_to_one')#one to one nahi hai
```
