```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[5,10,15,20,25,30]
y=[10,12,8,4,25,18]
plt.plot(x,y,color='red',marker='o',ms=10,mfc='yellow',mec='black',linestyle=':')#search mar lena baki
plt.show()
```

```python
years=[1950,1960,1970,1980,1990,2000,2010,2020]
company_x=[2.5,3.5,2.8,3.1,4.0,2.9,4.5,4.8]

years=[1950,1960,1970,1980,1990,2000,2010,2020]
company_y=[2.2,3.8,2.5,3.5,4.5,2.1,4.4,4.5]

font1={'family':'serif','color':'blue','size':14}
font2={'family':'serif','color':'brown','size':14}

plt.subplot(1,2,1)
plt.plot(years,company_x,color='blue',marker='o',ms=10,mfc='yellow',mec='black')
plt.grid(linestyle='dashed',color='red')
plt.xlabel('Years')
plt.ylabel('Company_x Capital')
plt.title('Company_x capital over years',fontdict=font1)

plt.subplot(1,2,2)
plt.plot(years,company_y,color='brown',marker='D',ms=10,mfc='black',mec='pink')
plt.grid(linestyle='dashed',color='hotpink')
plt.xlabel('Years')
plt.ylabel('Company_y Capital')
plt.title('Company_y capital over years',fontdict=font2)

plt.subplots_adjust(wspace=0.5)#MAST
plt.suptitle('Comparision of companies over years')
plt.show()
```

```python
years=[1950,1960,1970,1980,1990,2000,2010,2020]
pop_india=[33,38,50,60,75,80,120,140]
pop_china=[50,60,70,80,120,125,140,138]

plt.scatter(years,pop_india,color='red')
plt.scatter(years,pop_china,color='blue')

plt.grid(linestyle='dotted',color='black')
plt.show()
```

```python
country=['Russia','Canada','China','America','Brazil','Australian','India']
area=[6.7,6.2,5.6,4.5,3.8,3.2,2.4]

plt.barh(country,area,color='red',height=0.5)
#$plt.bar(country,area,color='red',width=0.5)
plt.xlabel('Countries')
plt.ylabel('Area per million square meter')
plt.title('Area comparision between countries')

plt.show()
```

```python
plt.subplot(1,2,1)
plt.bar(years,company_x,color='blue',width=1.0)
plt.subplot(1,2,2)
plt.bar(years,company_y,color='red',width=1.0)
plt.show()
```

```python
countries=['Russia','America','China','Canada','Indonasia','India','Pakistan','Brazil','England','bangladesh']
pop=[14.6,33.1,141,3.8,27.3,142,22.5,21.3,6.7,16.9]

#plt.scatter(countries,pop,color='red')
#plt.grid(linestyle='dotted',color='black')
#plt.show()

indices = range(len(countries))

plt.scatter(indices, pop, color='blue')
plt.xlabel('Countries')
plt.ylabel('Population in Crores')
plt.title('Population of Countries in Crores')
plt.xticks(indices, countries, rotation=45)
plt.grid(linestyle='dotted', color='red')
plt.show()
```
