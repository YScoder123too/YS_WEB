```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path1 = r"C:\Users\hp\Downloads\Delhi_Weather_Data.xlsx"
Delhi = pd.read_excel(file_path1)
Delhi
```

```python
file_path2 = r"C:\Users\hp\Downloads\Mumbai_Weather_Data.xlsx"
Mumbai = pd.read_excel(file_path2)
Mumbai
```

```python
file_path4 = r"C:\Users\hp\Downloads\Banglore.xlsx"
Bangalore = pd.read_excel(file_path4)
Bangalore
```

```python
file_path3 = r"C:\Users\hp\Downloads\Chennai.xlsx"
Chennai = pd.read_excel(file_path3)
Chennai
```

```python
font1 = {'family':'serif','color':'blue','size':14}
font2 = {'family':'serif','color':'darkred','size':14}

fig,ax = plt.subplots(2,2,figsize=(14,10))
plt.grid(linestyle = 'dashed')
ax[0,0].barh(Bangalore['Month'],Bangalore['Average Temperature (°C)'],color='magenta')
ax[0,0].set_xlabel("Average Temperature (°C)",fontdict=font1)
ax[0,0].set_ylabel("Months",fontdict=font1)

ax[0,1].plot(Bangalore['Month'],Bangalore['Average Wind Speed (m/s)'])
ax[0,1].set_xlabel("Months",fontdict=font2)
ax[0,1].set_ylabel("Wind Speed (km/h)",fontdict=font2)

ax[1,0].bar(Bangalore['Month'],Bangalore['Average Humidity (%)'])
ax[1,0].set_xlabel("Months",fontdict=font2)
ax[1,0].set_ylabel("Humidity (%)",fontdict=font2)

ax[1,1].plot(Bangalore['Month'],Bangalore['Average Air Pressure (hPa)'],color='magenta')
ax[1,1].set_xlabel("Months",fontdict=font1)
ax[1,1].set_ylabel("Air Pressure (hPa)",fontdict=font1)

plt.suptitle("Bangalore Temp",color='Darkred',size=20)
plt.show()
```
