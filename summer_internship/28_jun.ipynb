# Cleaning Data

Data Cleaning means fixes bad data in your data set.
Bad data could be:
Empty cells,
Data in wrong format,Wrong data,Duplicates

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("pets.csv")
df
```

```python
df1=df.dropna()#jisme koi value nahi hai vo drop ho jae #pura ud jaega lol..so beware
df1
```

```python
'''df2=df.dropna(inplace=True)
df2'''
```

```python
df2=df.fillna(50)
df2
```

```python
df['Height'].fillna(50,inplace=True)
df
```

```python
x=df['Weight'].mean()
print(x)
```

```python
df['Weight'].fillna(x,inplace=True)
df
```

```python
height_mean=df['Weight'].mean()
print(height_mean)
```

```python
weight_median=df['Weight'].median()
print(weight_median)
```

```python
df3=df[df['Weight'] == df['Weight'].max()]
df3
```

```python
df4=df[df['Weight'] == df['Weight'].min()]
df4
```

```python
df5=df[df['Date of Birth'] == df['Date of Birth'].min()]#sirf 1 ko dekh raha hai isliye
df5
```

```python
df5=df[df['Date of Birth'] == df['Date of Birth'].max()]
df5
```
