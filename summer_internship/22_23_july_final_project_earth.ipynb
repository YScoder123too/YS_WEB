```python
india_data=px.data.gapminder().query('country == "India"')
india_data
```

```python
#Task 4:Quick Visualization with bar chart
fig=px.bar(india_data,x='year',y='pop',height=400)
fig.show()
```

```python
fig=px.bar(india_data,x='year',y='pop',height=400,color='lifeExp',labels={'pop':'Population of India'})#lifeExp ek teesra factor hair through color
fig.show()
```

```python
gap2007=gapminder.query('year==2007')
fig=px.scatter(gap2007,x='gdpPercap',y='lifeExp',height=400,color='continent',size_max=60,size='pop',hover_data=['country'])
fig.show()
```

```python
gap2007=gapminder.query('year==2007')
fig=px.scatter(gap2007,x='gdpPercap',y='lifeExp',height=400,color='continent',size_max=60,size='pop',hover_data=['country']
              ,facet_col='continent'#individually continent vise ban gaya
              )
fig.show()
```

```python
gap2007=gapminder.query('year==2007')
fig=px.scatter(gapminder,x='gdpPercap',y='lifeExp',height=500,color='continent',size_max=40,size='pop',hover_name='country',
              animation_frame='year',log_x=True,animation_group='country',range_x=[25,10000],range_y=[25,90])
fig.show()
```

```python
gap2007=gapminder.query('year==2007')
fig=px.scatter(gapminder,x='gdpPercap',y='lifeExp',height=500,color='continent',size_max=40,size='pop',hover_name='country',
              animation_frame='year',log_x=True,animation_group='country',range_x=[25,10000],range_y=[25,90],
              labels=dict(pop='Population',gdpPercap='GDP Per Capita',lifeExp='Life Expectancy'))
fig.show()
```

```python
fig =px.scatter_geo(gapminder.query('year==2007'),locations='iso_alpha',color='continent',projection = 'orthographic',hover_name='country')
fig.show()
```

```python
fig =px.line_geo(gapminder.query('year==2007'),locations='iso_alpha',color='continent',projection = 'orthographic',hover_name='country')
fig.show()
```

```python
fig =px.scatter_geo(gapminder.query('year==2007'),locations='iso_alpha',color='continent',
                 size='pop',projection = 'orthographic',hover_name='country',height=500)
fig.show()
```

```python
fig =px.choropleth(gapminder,locations='iso_alpha',color='lifeExp',
                hover_name='country',height=500,animation_frame='year',
                   color_continuous_scale=px.colors.sequential.Plasma, projection = 'natural earth',)
fig.show()
```
