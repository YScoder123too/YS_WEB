```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = {
    'Month': ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'],
    'avg_temp': [14.6, 17.9, 23.6, 30.6, 34.9, 33.6, 30.6, 29.4, 29.3, 26.3, 20.1, 15.0],
    'Wind Speed': [7, 8, 10, 12, 14, 13, 11, 10, 8, 6, 6, 6],
    'Humidity': [80, 69, 59, 44, 42, 54, 74, 76, 70, 58, 58, 74],
    'Air Pressure': [1019, 1014, 1011, 1008, 1006, 1000, 1000, 1000, 1006, 1011, 1015, 1017]
}

# Create dataframes
delhi = pd.DataFrame(data)
mumbai = pd.DataFrame(data)

fig, bx = plt.subplots(4, 2, figsize=(15, 20))  # 4*2=8 graphs

font1 = {'family': 'serif', 'color': 'blue', 'size': 10}
font2 = {'family': 'serif', 'color': 'red', 'size': 10}
font3 = {'family': 'serif', 'color': 'green', 'size': 10}

# Delhi plots
bx[0, 0].plot(delhi['Month'], delhi['avg_temp'], color='r', marker='o', mfc='black', mec='pink', ms=10)
bx[0, 0].set_xlabel('Months', fontdict=font2)
bx[0, 0].set_ylabel('Average Temperature', fontdict=font2)

bx[0, 1].plot(delhi['Month'], delhi['Wind Speed'], color='r', marker='o', mfc='black', mec='pink', ms=10)
bx[0, 1].set_xlabel('Months', fontdict=font2)
bx[0, 1].set_ylabel('Wind Speed index', fontdict=font2)

bx[1, 0].plot(delhi['Month'], delhi['Humidity'], color='b', marker='*', mfc='black', mec='pink', ms=10)
bx[1, 0].set_xlabel('Months', fontdict=font1)
bx[1, 0].set_ylabel('Humidity Index', fontdict=font1)

bx[1, 1].plot(delhi['Month'], delhi['Air Pressure'], color='b', marker='*', mfc='black', mec='pink', ms=10)
bx[1, 1].set_xlabel('Months', fontdict=font1)
bx[1, 1].set_ylabel('Air Pressure Index', fontdict=font1)

# Mumbai plots
bx[2, 0].plot(mumbai['Month'], mumbai['avg_temp'], color='g', marker='>', mfc='black', mec='pink', ms=10)
bx[2, 0].set_xlabel('Months', fontdict=font3)
bx[2, 0].set_ylabel('Average Temperature', fontdict=font3)

bx[2, 1].plot(mumbai['Month'], mumbai['Wind Speed'], color='g', marker='>', mfc='black', mec='pink', ms=10)
bx[2, 1].set_xlabel('Months', fontdict=font3)
bx[2, 1].set_ylabel('Wind Speed index', fontdict=font3)

bx[3, 0].plot(mumbai['Month'], mumbai['Humidity'], color='g', marker='>', mfc='black', mec='pink', ms=10)
bx[3, 0].set_xlabel('Months', fontdict=font3)
bx[3, 0].set_ylabel('Humidity Index', fontdict=font3)

bx[3, 1].plot(mumbai['Month'], mumbai['Air Pressure'], color='g', marker='>', mfc='black', mec='pink', ms=10)
bx[3, 1].set_xlabel('Months', fontdict=font3)
bx[3, 1].set_ylabel('Air Pressure Index', fontdict=font3)

plt.suptitle('Comparison of Temperature between Delhi and Mumbai', color='Black', size=20)
# plt.xticks(rotation=90)

plt.subplots_adjust(wspace=0.5)  # MAST
plt.show()
```
