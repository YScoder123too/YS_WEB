```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
path= r"C:\Users\LENOVO\Downloads\student-alcohol-consumption.csv"
student_data=pd.read_csv(path)
student_data.head()
```

```python
sns.relplot(x='absences',y='G3',data=student_data,kind='scatter',color='blue')
plt.grid(linestyle='dashed',color='skyblue')
plt.show()
```

```python
sns.relplot(x='absences',y='G3',data=student_data,kind='line',color='blue')
plt.grid(linestyle='dashed',color='skyblue')
plt.show()
```

```python
sns.relplot(x='absences',y='G3',data=student_data,kind='scatter',color='blue',row='study_time')
plt.show()
```

```python
sns.relplot(x='absences',y='G3',data=student_data,kind='scatter',color='blue',col='study_time')
plt.show()
```

```python
sns.relplot(x='G1',y='G3',data=student_data,kind='scatter')
plt.show()
```

```python
mpg=pd.read_csv('mpg.csv')
mpg.head()
```

```python
sns.relplot(x='horsepower',y='mpg',data=mpg,kind='scatter',size='cylinders',hue='cylinders')#cylinders ek column hai
plt.show()
```

```python
sns.relplot(x='acceleration',y='mpg',data=mpg,kind='scatter',size='origin',hue='origin')#cylinders ek column hai#mast same naam likho
plt.show()
```

```python
sns.relplot(x='acceleration',y='mpg',data=mpg,kind='scatter',style='origin',hue='origin')#cylinders ek column hai#mast same naam likho
plt.show()
```

```python
sns.relplot(x='acceleration',y='mpg',data=mpg,kind='line',style='origin',hue='origin')#cylinders ek column hai#mast same naam likho
plt.show()
```

```python
sns.relplot(x='model_year',y='mpg',data=mpg,kind='line')#dark line mean hai,95% content is mean
plt.show()
```

```python
sns.relplot(x='model_year',y='mpg',data=mpg,kind='line',ci=None)#dark line mean hai,95% content is mean,light shadow bhag gayi
plt.show()
```

```python
sns.relplot(x='model_year',y='mpg',data=mpg,kind='line',ci='sd')#dark line mean hai,95% content is mean..ci=confidence interval
plt.show()
```

```python
sns.relplot(x='model_year',y='horsepower',data=mpg,kind='line',ci=None)#dark line mean hai,95% content is mean..ci=confidence interval
plt.show()
```

```python
sns.catplot(x='model_year',y='horsepower',data=mpg)
plt.show()
```
