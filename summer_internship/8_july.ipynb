```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path=r"C:\Users\LENOVO\Downloads\delhi.csv"
delhi=pd.read_csv(file_path)
delhi
```

```python
file_path=r"C:\Users\LENOVO\Downloads\mumbai.csv"
mumbai=pd.read_csv(file_path)
mumbai
```

```python
fig,ax=plt.subplots(3,2,figsize=(15,10))#3*2=6 graphs hongae.....(width,height)

font1={'family':'serif','color':'blue','size':10}
font2={'family':'serif','color':'red','size':10}
font3={'family':'serif','color':'green','size':10}

ax[0,0].plot(delhi['Month'],delhi['Minimum'],color='r',marker='o',mfc='black',mec='pink',ms=10)#delhi[x coordinate],delhi[y coordinate]
ax[0,0].set_xlabel('Months',fontdict=font2)
ax[0,0].set_ylabel('Minimum Temperature',font2)

ax[0,1].plot(mumbai['Month'],mumbai['Minimum'],color='r',marker='o',mfc='black',mec='pink',ms=10)
ax[0,1].set_xlabel('Months',font2)
ax[0,1].set_ylabel('Minimum Temperature',font2)

ax[1,0].plot(delhi['Month'],delhi['Maximum'],color='b',marker='*',mfc='black',mec='pink',ms=10)
ax[1,0].set_xlabel('Months',font1)
ax[1,0].set_ylabel('Maximum Temperature',font1)

ax[1,1].plot(mumbai['Month'],mumbai['Maximum'],color='b',marker='*',mfc='black',mec='pink',ms=10)
ax[1,1].set_xlabel('Months',font1)
ax[1,1].set_ylabel('Maximum Temperature',font1)

ax[2,0].plot(delhi['Month'],delhi['Average'],color='g',marker='>',mfc='black',mec='pink',ms=10)
ax[2,0].set_xlabel('Months',font3)
ax[2,0].set_ylabel('Average Temperature',font3)

ax[2,1].plot(mumbai['Month'],mumbai['Average'],color='g',marker='>',mfc='black',mec='pink',ms=10)
ax[2,1].set_xlabel('Months',font3)
ax[2,1].set_ylabel('Average Temperature',font3)

plt.suptitle('Comparision of temperature between Delhi and Mumbai',color='Black',size=20)
#plt.xticks(rotation=90)

plt.subplots_adjust(wspace=0.5)#MAST
plt.show()
```
