```python
year=[1950,1970,1990,2010]
population=[2.5,3.6,5.2,6.9]
plt.plot(year,population)
plt.show()
```

```python
plt.scatter(year,population)
plt.xlabel('Years')
plt.ylabel('Population in crores')
plt.title('Population Data')
plt.show()
```

```python
plt.bar(year,population)
plt.xlabel('Years')
plt.ylabel('Population in crores')
plt.title('Population Data')
plt.show()
```

```python
plt.barh(year,population)
plt.xlabel('Years')
plt.ylabel('Population in crores')
plt.title('Population Data')
plt.show()
```
