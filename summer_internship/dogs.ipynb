# Slicing DataFrame 

```python
#loc=>location,idoc=>index of document
df1=df.loc[:,['Dog Breed Group','Height','Weight','Life Span','Easy To Groom']]
'''[rows:columns],rows hanme already head se 5 select kara hai'''
df1.head()
```

```python
df2=df1.sort_values('Easy To Groom')#asending order
df2.head(200)
```
